Password Validation 

at least one captial letter

at least one number

at least on special character 

length 8-18

```python
import re 
def password_check(ps): 
    
    valid = "^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[~!@#$%^&*])[A-Za-z\d~!@#$%^&*]{8-18}$"

    #compile regex
    match_re = re.compile(valid)
    #searching regex
    res = re.search(match_re, ps) 
    if res: 
            print("Valid Password")
    else:
            print("Invalid Password")
        
def main():
    print("Please Enter password should be (1)One Capital letter \n(2)Special Letter \n(3)One Number \n(4)Length be 8-18 :")
    pwd = input()
    password_check(pwd)
           
    while pwd != 'n':
        pwd = input("You can try any password until enter")
        password_check(pwd)
        
main()
```
