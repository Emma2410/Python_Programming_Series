Regular Expression Functions 

findall = return the list containing all matches
search  = return the match object if there is a match
split = return list of string where it has been split
sub = replace one or more matches with a stirng


**[]**  a set of character

\        signals a special sequence(\d)

.       any character except new line

^       start with 

$       end with

*       zero or more occurences

?       zero or one  occurences

+       one or more occureance 

{}       exactly specified number of occurrences

|       either or (fall|stays}

()      capture of group


\A  - return the character if there is a 'characters' at the beginning of the string. "\AThe" - the is at the beginning of the sentence.

\d  -  Digit
\D  - does not contain digit
\s - retrun a match where white space character contain

\S - return a match, where there is no white space character


\w	Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character)

\W	Returns a match where the string DOES NOT contain any word characters	"\W"	

\Z	Returns a match if the specified characters are at the end of the string



```python
Functions 
findall() - return a list contain all match
search() - searches the string for a match
split() - return a list where the string has been split at each match 
sub() -  repalce the match with the text of your choice


The Match object has properties and methods used to retrieve information about the search, and the result:

.span() returns a tuple containing the start-, and end positions of the match.
.string returns the string passed into the function
.group() returns the part of the string where there was a match
```
