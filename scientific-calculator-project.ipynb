```python
def dig_to_bin(digit):
    items = []
    while digit > 0:
        mod = digit % 2
        items.append(mod)
        digit = digit // 2
    while items:
        print(items.pop()) 
        
def bin_to_dig(digit):
    dec_num = 0
    m = 1
    for i in digit:
        i= int(i)
        dec_num = dec_num + (i * m)
        m = m * 2
    print(dec_num)
    

def operation(f_digit,s_digit,opt):
    result = 0
    if opt == '+':
        result = f_digit + s_digit
    elif opt == '-':
        result = f_digit - s_digit 
    elif opt == '*':
        result = f_digit * s_digit 
    elif opt == '/':
        result = f_digit / s_digit 
    elif opt == '%':
        result = f_digit % s_digit
    else: 
        print("Wrong Operator")

    print("The answer is ", result)
    
def main():  
   

    chos = input("Type Scientific Calculator for \"s\" or Type Normal Calculator for \"n\"  ")
    if chos == "s":
        conv = input("Type Binary to Digital to \"bd\" \nType Digital to Binary to \"db\"")
        if conv == 'db':
            f_digit = int(input("Enter Decimal Number"))
            dig_to_bin(f_digit)
        elif conv == 'bd':
            f_digit = input("Enter Binary Number")
            bin_to_dig(f_digit)
    elif chos == "n":
        f_digit = int(input("Enter First Number"))
        s_digit = int(input("Enter Second Number"))
        opt = input("Enter Operator, +,-,*,/,%")
        operation(f_digit,s_digit,opt)
    else:
        print("Only Enter s and n")
    
main()
```
